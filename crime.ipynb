```python
clean_df = df.drop(columns=['Cross Street', 'Mocodes','Premis Cd','Rpt Dist No'])
clean_df
```

```python
# standardize 'Vict Sex' column to have consistent values: 'Male ', 'Female ', and 'Unspecified'
clean_df['Vict Sex'] = clean_df['Vict Sex'].apply(lambda x: 'Unspecified' if x != 'F' and x != 'M' else x)
clean_df.loc[clean_df['Vict Sex'] == 'M', 'Vict Sex'] = 'Male'
clean_df.loc[clean_df['Vict Sex'] == 'F', 'Vict Sex'] = 'Female'
```

```python
# check that all values changed as expected
valid = ['Male', 'Female', 'Unspecified']
(clean_df['Vict Sex'].isin(valid)).all()

check_df = pd.DataFrame({
    'original': df['Vict Sex'],
    'cleaned': clean_df['Vict Sex']
})

(clean_df.loc[clean_df['Vict Sex'] == 'M', 'Vict Sex'] == 'Male').all()
(clean_df.loc[clean_df['Vict Sex'] == 'F', 'Vict Sex'] == 'Female').all()
```

```python
df['Vict Descent'].unique()
```

```python

clean_df['Vict Descent'] = (
    clean_df['Vict Descent']
    .astype(str)      # convert everything to string
    .str.strip()      # remove leading/trailing whitespace
    .str.upper()      # make all text uppercase
)
# replace any value that is not a valid descent code with NaN
clean_df['Vict Descent'] = clean_df['Vict Descent'].replace('-', np.nan)
clean_df['Vict Descent'] = clean_df['Vict Descent'].replace(['-', 'X'], np.nan)
```

```python
descent_map = {
    'H': 'H: Hispanic/Latino',
    'W': 'W: White',
    'B': 'B: Black',
    'A': 'A: Other Asian',
    'C': 'C: Chinese',
    'J': 'J: Japanese',
    'K': 'K: Korean',
    'F': 'F: Filipino',
    'V': 'V: Vietnamese',
    'I': 'I: American Indian/Alaska Native',
    'P': 'P: Pacific Islander',
    'O': 'O: Other',
    'Z': 'Z: Asian Indian',
    'G': 'G: Guamanian',
    'U': 'U: Hawaiian',
    'D': 'D: Cambodian',
    'L': 'L: Laotian',
    'S': 'S: Samoan'
}
# replace '-' and 'X' with NaN then map the remaining values to their full descriptions, filling with 'Unspecified'
clean_df['Vict Descent'] = clean_df['Vict Descent'].map(descent_map).fillna('Unspecified')
```

```python
# function to get the most common crimes by their codes and descriptions
def get_top_crimes(df, top_n=10):
    return (df[['Crime Code', 'Crime Description']]  # select crime code and description
        .drop_duplicates()  # keep one row per crime code to avoid duplicate mappings
        .merge(
            df['Crime Code'].value_counts().rename('Count'),  # count number of occurrences per crime code
            left_on='Crime Code',
            right_index=True
        )
        .sort_values(by='Count', ascending=False)  # sort crimes by highest frequency
        .head(top_n)  # keep top N crimes
        .reset_index(drop=True) 
    )
    
get_top_crimes(clean_df)
```

```python
# function to get the most common area crimes happen by their codes and names
def get_top_crime_locations(df, top_n=10):
    return (df[['AREA', 'AREA NAME']]  # select area code and area name
        .drop_duplicates()  # keep one row per AREA to avoid duplicate mappings
        .merge(
            df['AREA'].value_counts().rename('Count'),  # count number of crimes per AREA
            left_on='AREA',
            right_index=True
        )
        .sort_values(by='Count', ascending=False)  # sort areas by highest crime count
        .head(top_n)  # keep top N areas
        .reset_index(drop=True)  
    )
    
get_top_crime_locations(clean_df)
```

```python
def get_crimes_by_month_name(df):
    return (
        df.groupby(df['Date Occurred'].dt.month_name())  # group by month name from Date Occurred
        .size()  # count number of crimes per month
        .reindex(['January', 'February', 'March', 'April', 'May', 'June',
            'July', 'August', 'September', 'October', 'November', 'December'
        ])  # reorder months into correct calendar order
        .reset_index(name='Count')  # convert into DataFrame with Count column
        .rename(columns={'Date Occurred': 'Month'})  # rename column for clarity
    )

get_crimes_by_month_name(clean_df)
```

```python
# double check that the month names are correct by comparing to value counts of month numbers
clean_df['Date Occurred'].dt.month_name().value_counts()
```

```python
# function to get number of crimes per year based on Date Occurred
def get_crimes_by_year(df):
    return (
        df.groupby(df['Date Occurred'].dt.year)  # group by year extracted from Date Occurred
        .size()  # count number of crimes per year
        .reset_index(name='Count')  # convert to DataFrame with Count column
        .rename(columns={'Date Occurred': 'Year'})  # rename column for clarity
        .sort_values(by='Year')  # sort years in chronological order
    )

get_crimes_by_year(clean_df)
```

```python
# function to get number of crimes per hour sorted by frequency
def get_crimes_by_hour(df):
    return (
        df.groupby(pd.to_datetime(df['Time Occurred'].astype(str)).dt.hour)  # group by hour
        .size()  # count crimes per hour
        .reset_index(name='Count')  # convert to DataFrame with Count column
        .rename(columns={'Time Occurred': 'Hour'})  # rename column
        .sort_values(by='Count', ascending=False)  # sort by highest count
        .reset_index(drop=True)  
    )

get_crimes_by_hour(clean_df)
```
