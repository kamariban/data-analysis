## Analytical Functions

Custom functions created to extract key patterns from the dataset. They allow for reusable analysis and make it easier to explore relationships between time, location, and crime type.

- Top Crimes: Identifies the most frequent crime types based on code and description
- Top Locations: Determines which areas experience the highest number of crimes
- Crimes by Month: Highlights seasonal patterns in crime 
- Crimes by Hour: Examines how crime frequency can change in 1 day


```python
# function to get the most common crimes by their codes and descriptions
def get_top_crimes(df, top_n=10):
    return (df[['Crime Code', 'Crime Description']]  # select crime code and description
        .drop_duplicates()  # keep one row per crime code to drop duplicates
        .merge(
            df['Crime Code'].value_counts().rename('Count'),  # count number of instances per crime code
            left_on='Crime Code',
            right_index=True
        )
        .sort_values(by='Count', ascending=False)  # sort crimes by highest frequency
        .head(top_n)  # keep top N crimes
        .reset_index(drop=True) 
    )
    
get_top_crimes(clean_df)
```

```python
# function to get the most common area crimes happen by their codes and names
def get_top_crime_locations(df, top_n=10):
    return (df[['AREA', 'AREA NAME']]  # select area code and area name
        .drop_duplicates()  # keep one row per AREA to avoid duplicates
        .merge(
            df['AREA'].value_counts().rename('Count'),  # count number of crimes per AREA
            left_on='AREA',
            right_index=True
        )
        .sort_values(by='Count', ascending=False)  # sort areas by highest crime count
        .head(top_n)  # keep top N areas
        .reset_index(drop=True)  
    )
    
get_top_crime_locations(clean_df)
```

```python
def get_crimes_by_month_name(df):
    return (
        df.groupby(df['Date Occurred'].dt.month_name())  # group by month name from Date Occurred
        .size()  # count number of crimes per month
        .reindex(['January', 'February', 'March', 'April', 'May', 'June',
            'July', 'August', 'September', 'October', 'November', 'December'
        ])  # reorder months into correct order
        .reset_index(name='Count')  # convert into DataFrame with Count column
        .rename(columns={'Date Occurred': 'Month'})  # rename column 
    )

get_crimes_by_month_name(clean_df)
```

```python
# function to get number of crimes per hour sorted by frequency
def get_crimes_by_hour(df):
    return (
        df.groupby(pd.to_datetime(df['Time Occurred'].astype(str)).dt.hour)  # group by hour
        .size()  # count crimes per hour
        .reset_index(name='Count')  # convert to DataFrame with Count column
        .rename(columns={'Time Occurred': 'Hour'})  # rename column
        .sort_values(by='Count', ascending=False)  # sort by highest count
        .reset_index(drop=True)  
    )

get_crimes_by_hour(clean_df)
```

## Crime Timing Patterns

- This analysis explores whether certain crimes peak at any specific times of day.

Conclusion: 
- Vehicle theft makes up the majority of crimes, with most of them happening at night, 6 pm being the peak, but reports continue throughout the night at around a similar level of occurrence. The interesting outliers were identity theft, which mostly occurred at around noon and midnight, with the noon occurrences being the single most reported crime in the dataset. The increase in car theft at night speaks for itself, but the identity theft discovery led me to some research where I learned that because noon is when many people are often distracted by work and lunch hours, this is a peak time for the use of public Wi-Fi or ATMs, which increases vulnerability to scams and card skimming. At midnight, online activity shifts to more automated processes, so criminals try to take advantage of this and use bots to test stolen credentials, run phishing attempts, and make unauthorized transactions while victims are less likely to notice.

```python

# function to get number of crimes by hour and crime type
def crime_by_hour(df, top_n=10):
    return (
        df.groupby([
            pd.to_datetime(df['Time Occurred'].astype(str)).dt.hour,  # convert and extract hour
            'Crime Description'
        ])
        .size()  # count occurrences
        .reset_index(name='Count')  # convert to DataFrame
        .sort_values(by='Count', ascending=False)  # sort by highest frequency
        .head(top_n)  # keep top N combinations
        .reset_index(drop=True)  # clean index
    )
    
crime_by_hour(clean_df)

    
```

```python
# function to get number of crimes by month and crime type
def crime_by_month(df, top_n=10):
    return (
        df.groupby([
            pd.to_datetime(df['Date Occurred'].astype(str)).dt.month,  # convert and extract month
            'Crime Description'
        ])
        .size()  # count occurrences
        .reset_index(name='Count')  # convert to DataFrame
        .sort_values(by='Count', ascending=False)  # sort by highest frequency
        .head(top_n)  # keep top N combinations
        .reset_index(drop=True)  # clean index
    )
    
crime_by_month(clean_df)
```

## Crime Reporting Delays

- This analysis examines how long it takes for crimes to be reported after they occur. A report delay variable was created by calculating the difference between the date reported and the date occurred.
- This allows for identification of crimes that are reported significantly later and helps determine whether reporting delays do vary across different crime types.

Conclusion : 
- Most crimes are reported either the day of or the day after, with same-day reporting as the clear majority across almost every crime type. This suggests that most incidents are immediately noticed or, because of the severity of the crime, require immediate police involvement. And this can be seen in the data for the more visible and violent crimes, such as assault and burglary. Vehicle-related crimes were the repeating outlier, appearing with a one-day delay for a few crimes, which is likely because in these crimes the theft or damage is only discovered later, often the morning after the incident, which coincides with the pattern seen in my earlier crime-by-hour analysis, with vehicle theft peaking at night. Overall, reporting delays were surprisingly minimal, which is good. This tells us that victims are generally aware of crimes quickly and take action. 



```python
# function to get number of crimes by report delay and crime type
def report_delay(df, top_n=10):
    delay = (
        pd.to_datetime(df['Date Reported']) - 
        pd.to_datetime(df['Date Occurred'])
    ).dt.days  # calculate delay in days

    return (
        df.assign(Report_Delay=delay)  # create new column
        .groupby(['Report_Delay', 'Crime Description'])  # group by delay and crime
        .size()
        .reset_index(name='Count')  # clean DataFrame
        .sort_values(by='Count', ascending=False)
        .head(top_n)
        .reset_index(drop=True)
    )
report_delay(clean_df, 20)
```
