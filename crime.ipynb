```python
clean_df = df.drop(columns=['Cross Street', 'Mocodes','Premis Cd','Rpt Dist No'])
clean_df
```

```python
clean_df.info()
```

```python
# standardize 'Vict Sex' column to have consistent values: 'Male ', 'Female ', and 'Unspecified'
clean_df['Vict Sex'] = clean_df['Vict Sex'].apply(lambda x: 'Unspecified' if x != 'F' and x != 'M' else x)
clean_df.loc[clean_df['Vict Sex'] == 'M', 'Vict Sex'] = 'Male'
clean_df.loc[clean_df['Vict Sex'] == 'F', 'Vict Sex'] = 'Female'
```

```python
# check that all values changed as expected
valid = ['Male', 'Female', 'Unspecified']
(clean_df['Vict Sex'].isin(valid)).all()

check_df = pd.DataFrame({
    'original': df['Vict Sex'],
    'cleaned': clean_df['Vict Sex']
})

(clean_df.loc[clean_df['Vict Sex'] == 'M', 'Vict Sex'] == 'Male').all()
(clean_df.loc[clean_df['Vict Sex'] == 'F', 'Vict Sex'] == 'Female').all()
```

```python
df['Vict Descent'].unique()
```

```python

clean_df['Vict Descent'] = (
    clean_df['Vict Descent']
    .astype(str)      # convert everything to string
    .str.strip()      # remove leading/trailing whitespace
    .str.upper()      # make all text uppercase
)
# replace any value that is not a valid descent code with NaN
clean_df['Vict Descent'] = clean_df['Vict Descent'].replace('-', np.nan)
clean_df['Vict Descent'] = clean_df['Vict Descent'].replace(['-', 'X'], np.nan)
```

```python
descent_map = {
    'H': 'H: Hispanic/Latino',
    'W': 'W: White',
    'B': 'B: Black',
    'A': 'A: Other Asian',
    'C': 'C: Chinese',
    'J': 'J: Japanese',
    'K': 'K: Korean',
    'F': 'F: Filipino',
    'V': 'V: Vietnamese',
    'I': 'I: American Indian/Alaska Native',
    'P': 'P: Pacific Islander',
    'O': 'O: Other',
    'Z': 'Z: Asian Indian',
    'G': 'G: Guamanian',
    'U': 'U: Hawaiian',
    'D': 'D: Cambodian',
    'L': 'L: Laotian',
    'S': 'S: Samoan'
}
# replace '-' and 'X' with NaN, then map the remaining values to their full descriptions, filling any unmapped values with 'Unspecified'

clean_df['Vict Descent'] = clean_df['Vict Descent'].map(descent_map).fillna('Unspecified')
```

```python
# remove impossible 0 and negative ages replace with NaN
clean_df.loc[clean_df['Vict Age'] <= 0, 'Vict Age'] = np.nan
```

```python
# double check that all valid ages are unchanged
(clean_df.loc[df['Vict Age'] > 0, 'Vict Age'] == df.loc[df['Vict Age'] > 0, 'Vict Age']).all()
```

```python
# convert to datetime
clean_df['Date Rptd'] = pd.to_datetime(clean_df['Date Rptd'], errors='coerce')
clean_df['DATE OCC'] = pd.to_datetime(clean_df['DATE OCC'], errors='coerce')
```

```python
clean_df[['Date Rptd', 'DATE OCC']]
```

```python
(clean_df['TIME OCC'] <= 2359).all()
```

```python
clean_df['TIME OCC'] = (
    pd.to_datetime(
        clean_df['TIME OCC']
        .astype(str)      # convert integers to strings
        .str.zfill(4),     # pad with leading zeros : 100 → 0100
        format='%H%M',     # specify HHMM time format
        errors='coerce'    # invalid values → NaT
    )
    .dt.time              # extract time 
)
```

```python
(df.index == clean_df.index).all()
```

```python
# verify hours matches original TIME OCC (HHMM → hour)
(
    (df['TIME OCC'] // 100) ==
    pd.to_datetime(clean_df['TIME OCC'].astype(str)).dt.hour
).all()
```
